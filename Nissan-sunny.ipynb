```python
temp = [lowreviews.replace('...', ' ').replace('\n', ' ')]
```

```python
temp[0] = str.lower(temp[0])
```

```python
temp
```

```python
tokens = []
stops = ['hello' , 'everyone', 'looks', 'car', 'honda', 'city' , 'going', 'hi','friends']
tokens.append([w for w in word_tokenize(temp[0]) if w.isalpha() and w not in stops and w not in stopwords.words('english')])
```

```python
len(tokens[0])
```

```python
 t = Counter(tokens[0]).most_common(500)
```

```python
var = []
for i in range(len(t)):
    var.append(t[i][0])
```

```python
string = ' '.join(var)
```

```python
string
```

```python
mask = np.array(Image.open('D:/imafe.png'))
wc = WordCloud(background_color="white", max_words=1000, mask=mask,
               max_font_size=90, random_state=42)
wc.generate(string)

image_colors = ImageColorGenerator(mask)
plt.figure(figsize=[7,7])
plt.imshow(wc.recolor(color_func=image_colors), interpolation="bilinear")
plt.axis("off")
plt.show()
```
