```python
temp[0] = str.lower(temp[0])
```

```python
from nltk.corpus import stopwords
```

```python
from nltk.tokenize import word_tokenize
from collections import Counter
tokens = []
stops = ['hello' , 'everyone', 'looks', 'car', 'honda', 'city' , 'going', 'hi','friends']
tokens.append([w for w in word_tokenize(temp[0]) if w.isalpha() and w not in stops and w not in stopwords.words('english')])
```

```python
len(tokens[0])
```

```python
 t = Counter(tokens[0]).most_common(500)
```

```python
var = []
for i in range(len(t)):
    var.append(t[i][0])
```

```python
string = ' '.join(var)
```
